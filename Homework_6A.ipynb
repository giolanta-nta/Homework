```python
import numpy as np
df.isnull().sum()
```

```python
df['age'] = df['age'].replace({np.nan: df['age'].mean()})
df['age'].isnull().sum()
```

```python
df['age'] = df['age'].replace({-42: df['age'].mean()})
df['age'] = df['age'].replace({0.0: df['age'].mean()})
df['age'] = df['age'].replace({-7: df['age'].mean()})
df['age'] = df['age'].replace({12: df['age'].mean()})
```

```python
#Πόσο χρονών είναι ο μέσος billionaire? Πόσο χρονών είναι οι self made vs. non self made?
round(df['age'].mean(), 2)
```

```python
round(df['age'].median(), 2)
```

```python
round(df[df['selfmade'] == 'self-made']['age'].mean(), 2)
```

```python
round(df[df['selfmade'] != 'selfmade']['age'].mean(), 2)
```

```python
#Ποιοι είναι οι νεότεροι billionaires? Ποιοι οι γηραιότεροι? Φτιάξτε ένα γράφημα με τη διασπορά των ηλικιών.
df['age'].sort_values(ascending=True)
```
