```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,mean_squared_error,r2_score
```

```python
df = pd.read_csv("/test_8i3B3FC.csv")
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.head()
```

```python
x = df[['Tag', 'Reputation', 'Answers']]
# Μετατροπή object σε integer/ float: pd.get_dummies(x, columns=['Tag']
#Αποφυγή πολυσυγγραμικότητας: drop_first=True
x = pd.get_dummies(x, columns=['Tag'], drop_first=True)
y = df[['Views']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.8, random_state=50)
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
plt.scatter(y_test, y_pred)
```

```python
print(y_pred[:5])
```

```python
print(model.intercept_)
```

```python
print(model.coef_)
```

```python
mse=mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```
