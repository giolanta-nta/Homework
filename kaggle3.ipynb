```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,mean_squared_error,r2_score
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv ("/advertising.csv")
```

```python
display(df.head())
```

```python
df.shape
```

```python
df.dtypes
```

```python
x=df[['TV','Radio','Newspaper']]
y=df['Sales']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.8,random_state=48)
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
plt.scatter(y_test,y_pred)
```

```python
print(y_pred[:5])
```

```python
mse=mean_squared_error(y_test, y_pred)
r2_score=r2_score(y_test, y_pred)
```

```python
print(mse)
print(r2_score)
```

```python
sample=[[230,37,69]]
prediction=model.predict(sample)
print("Prediction",prediction)
```

```python
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")

# Assuming the order of features is 'TV', 'Radio', 'Newspaper'
feature_names = x.columns
print("\nLinear Regression Equation:")
equation = f"Sales = {model.intercept_:.2f}"
for i, coef in enumerate(model.coef_):
    equation += f" + ({coef:.2f} * {feature_names[i]})"
print(equation)
```
